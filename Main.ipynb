# Establishing a Connection to the Sakila Database

```python
from sqlalchemy import create_engine

# Replace with your actual database connection string
engine = create_engine('mysql+pymysql://username:password@host/dbname')
```

Function rentals_month

This function retrieves rental data for a specific month and year from the Sakila database:




```python
import pandas as pd

def rentals_month(engine, month, year):
    query = f"""
    SELECT rental_id, rental_date, inventory_id, customer_id
    FROM rental
    WHERE MONTH(rental_date) = {month} AND YEAR(rental_date) = {year}
    """
    with engine.connect() as connection:
        df = pd.read_sql(query, connection)
    return df
```

Function rental_count_month

This function calculates the number of rentals made by each customer during a given month and year:

```python
def rental_count_month(df, month, year):
    rental_count = df.groupby('customer_id').size().reset_index(name=f'rentals_{month:02}_{year}')
    return rental_count
```

Function compare_rentals

This function compares the rental activity between two different months and returns the differences

```python
def compare_rentals(df1, df2):
    comparison_df = pd.merge(df1, df2, on='customer_id', how='outer', suffixes=('_month1', '_month2')).fillna(0)
    comparison_df['difference'] = comparison_df.iloc[:, 1] - comparison_df.iloc[:, 2]
    return comparison_df
```

Putting It All Together

You can now retrieve the rental data for May and June, count the rentals, and compare the activity between the two months:

```python
engine = create_engine('mysql+pymysql://username:password@hostname/dbname')
```

```python
# Retrieve data for May 2005
may_rentals = rentals_month(engine, 5, 2005)
may_count = rental_count_month(may_rentals, 5, 2005)

# Retrieve data for June 2005
june_rentals = rentals_month(engine, 6, 2005)
june_count = rental_count_month(june_rentals, 6, 2005)

# Compare the rental activity between May and June
rental_comparison = compare_rentals(may_count, june_count)

print(rental_comparison)
```
